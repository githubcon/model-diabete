4) Pour normaliser chaque colonne, exécuter le code suivant: data_norm=(data[colonne]-moyenne)/ecart_type

```python
data_norm = diabete_population.copy()

for i in data_norm.iloc[:,0:7]:
    data_norm[i] = (data_norm[i] - data_norm[i].mean())/data_norm[i].std()
data_norm
```

```python
# Autre méthode de normalisation
data_norm = diabete_population.copy()

data_norm["grossesses"] =  (diabete_population['grossesses'] - diabete_population['grossesses'].mean()) /  diabete_population['grossesses'].std()
data_norm["glucose"] =  (diabete_population['glucose'] - diabete_population['glucose'].mean()) /  diabete_population['glucose'].std()
data_norm["pression"] =  (diabete_population['pression'] - diabete_population['pression'].mean()) /  diabete_population['pression'].std()
data_norm["insuline"] =  (diabete_population['insuline'] - diabete_population['insuline'].mean()) /  diabete_population['insuline'].std()
data_norm["imc"] =  (diabete_population['imc'] - diabete_population['imc'].mean()) /  diabete_population['imc'].std()
data_norm["K"] =  (diabete_population['K'] - diabete_population['K'].mean()) /  diabete_population['K'].std()
data_norm["age"] =  (diabete_population['age'] - diabete_population['age'].mean()) /  diabete_population['age'].std()

data_norm
```

12) Considérons un nouve individu, dans chaque représentation (1 et 2), les paramètres de normalisés de cet individu consitueron les coordonnées d'un point.

On observant ses 11 plus proches voisins, on souhaite prédire si l'individu est malade ou pas

- Si la majorité de ses voisins est malade, alors on prédirera qu'il est malade
- Si la majorité de ses voisins est non malade alors on prédirera qu'il n'estpas malade.
- Parmi les deux représenation laquelle favorise au mieux la prédiction?

. La deuxième representation favorise mieux la prédiction

dist(insulline,glucose,grossesses) = racine carrée(coordonnées insulines*2 + coordonnées glucose*2 + coordonnées grossesses*2)

```python
data_norm
```

```python
diabete_patients_norm
```

```python
dist = []
for i in range(len(data_norm)):
    grossesses = (diabete_patients_norm.iloc[0][0] - data_norm.iloc[i][0])**2
    glucose = (diabete_patients_norm.iloc[0][1] - data_norm.iloc[i][1])**2
    pression = (diabete_patients_norm.iloc[0][2] - data_norm.iloc[i][2])**2
    calcul = ( grossesses +  glucose +  pression)**0.5
    dist.append(calcul)
dist    
```

```python
len(dist)
```

```python
data_norm["dist"] = dist
data_norm
```

```python
res=data_norm.sort_values(by="dist").iloc[0:12,:]
res['M'].sum()
```

```python
# Calcul pour le deuxième patient

dist1 = []

for i in range(len(data_norm)):
    grossesses = (diabete_patients_norm.iloc[1][0] - data_norm.iloc[i][0])**2
    glucose = (diabete_patients_norm.iloc[1][1] - data_norm.iloc[i][1])**2
    pression = (diabete_patients_norm.iloc[1][2] - data_norm.iloc[i][2])**2
    calcul = ( grossesses +  glucose +  pression)**0.5
    dist1.append(calcul)
dist1 
```

```python
data_norm["dist1"] = dist1
data_norm
```

```python
res1=data_norm.sort_values(by="dist1").iloc[0:12,:]
res1['M'].sum()
```

```python
data_norm
```

```python
data_norm 

for  pat in range(len(diabete_patients_norm)):
    gluc_pat=diabete_patients_norm.iloc[pat, 1]
    pression_pat=diabete_patients_norm.iloc[pat, 2]
    grossesse_pat=diabete_patients_norm.iloc[pat, 0]
    distances=[]
    
    for i in range(len(data_norm)):
        gluc_pop=data_norm.iloc[i, 1]
        pression_pop=data_norm.iloc[i, 2]
        grossesse_pop=data_norm.iloc[i, 0]
        
        glucose = (gluc_pat - gluc_pop)**2
        pression = (pression_pat - pression_pop)**2
        grossesse = (grossesse_pat - grossesse_pop)**2
        calcul = ( grossesse +  glucose +  pression)**0.5
        distances.append(calcul)
        
    data_norm[pat]=distances      
```
